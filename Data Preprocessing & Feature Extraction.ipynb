# Extracting Resolution Features

```python
splitted = df['ScreenResolution'].str.split('x',expand=True,n=1)
splitted
```

```python
df['x_resolution'] = splitted[0]
df['y_resolution'] = splitted[1]
```

```python
df.head(3)
```

```python
# using regex expression for finding pattern of nnnn in string x
df.x_resolution = df['x_resolution'].str.replace(',','').str.findall(r'(\d+\.?\d+)').apply(lambda x: x[0])
```

```python
df['x_resolution'] = df['x_resolution'].astype('int32')
df['y_resolution'] = df['y_resolution'].astype('int32')
```

```python
# ppi is a feature that may replace inches, x_resolution and y_resolution
# ppi = x**2 + y**2 / inches

df['PPI'] = (((df['x_resolution']**2 + df['y_resolution']**2)**0.5) / df['Inches']).astype('float')
```

```python
df.drop(columns=['x_resolution','y_resolution','ScreenResolution','Inches'],inplace=True)
```

# Extracing CPU Features

```python

def getInitials(x):
    return " ".join(x.split()[0:3])

df['CPUF'] = df['Cpu'].apply(getInitials)
```

```python
def getProcessorFeature(x):
    if(x == 'Intel Core i5' or x == 'Intel Core i7' or x == 'Intel Core i3'):
        return x
    elif(x.split()[0] == "Intel"):
        return "Intel Other"
    else:
        return "AMD"
    
```

```python
df.sample(30)
```

```python
df['Processor'] = df['CPUF'].apply(getProcessorFeature)
```

```python
df.drop(columns=['CPUF'],inplace=True)
```

# Memory Feature Extraction

```python
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '')
df["Memory"] = df["Memory"].str.replace('TB', '000')
new = df["Memory"].str.split("+", n = 1, expand = True)
```

```python
df["firstMemory"]= new[0]
df["firstMemory"]=df["firstMemory"].str.strip()
df["secondMemory"]= new[1]
df.sample(5)
```

```python

df["Layer1HDD"] = df["firstMemory"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["firstMemory"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["firstMemory"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["firstMemory"].apply(lambda x: 1 if "Flash Storage" in x else 0)
df.sample(4)
```

```python
df['firstMemory'] = df['firstMemory'].str.replace(r'\D', '')

df["secondMemory"].fillna("0", inplace = True)

df["Layer2HDD"] = df["secondMemory"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["secondMemory"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["secondMemory"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["secondMemory"].apply(lambda x: 1 if "Flash Storage" in x else 0)
```

```python

df['secondMemory'] = df['secondMemory'].str.replace(r'\D', '')

df["firstMemory"] = df["firstMemory"].astype(int)
df["secondMemory"] = df["secondMemory"].astype(int)


df["HDD"]=(df["firstMemory"]*df["Layer1HDD"]+df["secondMemory"]*df["Layer2HDD"])
df["SSD"]=(df["firstMemory"]*df["Layer1SSD"]+df["secondMemory"]*df["Layer2SSD"])
df["Hybrid"]=(df["firstMemory"]*df["Layer1Hybrid"]+df["secondMemory"]*df["Layer2Hybrid"])
df["Flash_Storage"]=(df["firstMemory"]*df["Layer1Flash_Storage"]+df["secondMemory"]*df["Layer2Flash_Storage"])

df.drop(columns=['firstMemory', 'secondMemory', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)

```
